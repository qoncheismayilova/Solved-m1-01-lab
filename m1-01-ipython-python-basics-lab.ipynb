#### Task 4: 
Apply control flow to a small record list
Create a list of at least 12 dictionaries representing user activity events with keys user_id, event_type, and duration_seconds. Include at least two records with invalid duration_seconds values (such as negative numbers or non-numeric strings). Write a loop that builds a new list called cleaned_events containing only valid records and adds a new key duration_minutes to each cleaned record. After the loop, verify two things: the count of cleaned_events matches the number of valid records you expect, and every cleaned record contains all required keys. Inspect two cleaned records in IPython to confirm the transformation.

```python
events = [
    {"user_id": 1, "event_type": "login", "duration_seconds": 120},
    {"user_id": 2, "event_type": "click", "duration_seconds": 45},
    {"user_id": 3, "event_type": "logout", "duration_seconds": "NaN"},  # etibarsız
    {"user_id": 4, "event_type": "login", "duration_seconds": 60},
    {"user_id": 5, "event_type": "click", "duration_seconds": -10},     # etibarsız
    {"user_id": 6, "event_type": "purchase", "duration_seconds": 300},
    {"user_id": 7, "event_type": "click", "duration_seconds": 30},
    {"user_id": 8, "event_type": "login", "duration_seconds": 150},
    {"user_id": 9, "event_type": "logout", "duration_seconds": 20},
    {"user_id": 10, "event_type": "click", "duration_seconds": 40},
    {"user_id": 11, "event_type": "purchase", "duration_seconds": 500},
    {"user_id": 12, "event_type": "login", "duration_seconds": 90},
]

events
```

```python
cleaned_events = []

for record in events:
    duration = record["duration_seconds"]
    if isinstance(duration, (int, float)) and duration >= 0:
        cleaned_record = record.copy()
        cleaned_record["duration_minutes"] = duration / 60
        cleaned_events.append(cleaned_record)
print("Count of cleaned event:",len(cleaned_events))
#Count of cleaned event: 10
```

```python
print(cleaned_events[0] )
print(cleaned_events[4])
#{'user_id': 1, 'event_type': 'login', 'duration_seconds': 120, 'duration_minutes': 2.0}
{'user_id': 7, 'event_type': 'click', 'duration_seconds': 30, 'duration_minutes': 0.5}
```

#### Task5
Summarize and capture a small output
Compute the following summaries using only core Python structures. First, build a dictionary of event counts by event_type. Second, build a dictionary of average duration_minutes by event_type. Third, compute the number of unique users using a set. After computing the summaries, validate one result by recomputing it in a different way, such as summing the event counts and comparing to len(cleaned_events).

Next, build a CSV-formatted string with three columns: metric, key, and value. Include one row per event type for the counts, one row per event type for the averages, and a final row for the unique user count. If you choose to save the string to disk while working, read it back and confirm one value matches your in-memory results. Include the final CSV string in your submission file.

```python
event_counts = {}

for record in cleaned_events:
    etype = record["event_type"]
    event_counts[etype] = event_counts.get(etype, 0) + 1

print("Event counts:", event_counts)
#Event counts: {'login': 4, 'click': 3, 'purchase': 2, 'logout': 1}
```

```python
duration_sums = {}
duration_counts = {}
for record in cleaned_events:
    etype = record["event_type"]
    duration = record["duration_minutes"]
    duration_sums[etype] = duration_sums.get(etype, 0) + duration
    duration_counts[etype] = duration_counts.get(etype, 0) + 1
avg_duration = {etype: duration_sums[etype]/duration_counts[etype] for etype in duration_sums}

print("Average duration_minutes:", avg_duration)
#Average duration_minutes: {'login': 1.75, 'click': 0.6388888888888888, 'purchase': 6.666666666666667, 'logout': 0.3333333333333333}
```

```python
unique_users = {record["user_id"] for record in cleaned_events}
num_unique_users = len(unique_users)

print("Unique users:", num_unique_users)
#Unique users: 10
```

```python
total_count = sum(event_counts.values())
print("Total event count matches cleaned_events:", total_count == len(cleaned_events))
#Total event count matches cleaned_events: True
```

```python
csv_lines = ["metric,key,value"]  
for etype, count in event_counts.items():
 csv_lines.append(f"count,{etype},{count}")

for etype, avg in avg_duration.items():
    csv_lines.append(f"average_duration,{etype},{avg:.2f}")
csv_lines.append(f"unique_users,all,{num_unique_users}")
csv_string = "\n".join(csv_lines)

print(csv_string)
```
